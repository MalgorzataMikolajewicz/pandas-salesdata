### Changing date format and adding new time periods

```python
merged_ALL['Time period']
```

```python
from datetime import datetime, date
```

```python
def substitute1(text):
    text = text[0:8]
    return text
```

```python
merged_ALL['Time period'] = merged_ALL['Time period'].apply(substitute1)
```

```python
merged_ALL['Time period'] = merged_ALL['Time period'].apply(pd.Timestamp)
```

```python
unique_months = sorted(merged_ALL['Time period'].unique(), reverse=True)
```

```python
lp = []
mat = []
n = 0
m = 0
for month in unique_months:
    n += 1
    lp.append(n)
for l in lp:
    if l/12 <= m:
        mat.append("MAT{}".format(m-1))
    else:
        m += 1
        mat.append("MAT{}".format(m-1))
```

```python
lp = []
r_quater = []
n = 0
for month in unique_months:
    n += 1
    lp.append(n)
q = [1,1,1,2,2,2,3,3,3,4,4,4]*(len(lp)//12)
for l in lp:
    r_quater.append("RQ{}".format(q[l-1]))
```

```python
year = []
for month in pd.DatetimeIndex(unique_months):
    year.append(month.year)
```

```python
quater = []
for month in pd.DatetimeIndex(unique_months):
    quater.append("Q{}".format((month.month-1)//3+1))
```

```python
half = []
for month in pd.DatetimeIndex(unique_months):
    half.append("H{}".format(month.month//7+1))
```

```python
ytd = []
lastdata = pd.DatetimeIndex(unique_months)[0].year
for month in pd.DatetimeIndex(unique_months):
    ytd.append("YTD{}".format(lastdata-month.year))
```

```python
dict_mat = pd.DataFrame({'Time period': unique_months, 'MAT': mat, 'YEAR': year, 'QUATER': quater, 'HALF': half, 'ROLLING QUATER': r_quater, 'YTD': ytd})
```

```python
merged_ALL_WITH_MAT = pd.merge(merged_ALL, dict_mat, on=['Time period'])
```

```python
merged_ALL_WITH_MAT.rename(columns={'Time period':'Month'}, inplace=True)
```
